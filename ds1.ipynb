```python
import pandas as pd
from IPython.display import display
file=pd.read_csv('moviesdatabase.txt',sep="\,",names=['Movie_ID','Movie_name','Year_of_release','Rating_of_the_movie','Movie_duration_in_seconds'])
display(file)
```

```python
total_rows=len(file)
display('Total number of movies')
display(total_rows+1)
```

```python
rating=file[file['Rating_of_the_movie']>4]
display(rating)
```

```python
release=file[(file['Year_of_release']>1950) & (file['Year_of_release']<1960)]
display(release)
```

```python
rating=file[(file['Rating_of_the_movie']>3) & (file['Rating_of_the_movie']<4)]
display(rating)
```

```python
timing=file[file['Movie_duration_in_seconds']>7200]
display(timing)
```

```python
eachyear=file.groupby(['Year_of_release']).count()
print(eachyear)
```
